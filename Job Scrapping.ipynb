```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

url1 = 'https://www.linkedin.com/jobs/search?keywords=Marketing%20Data%20Analyst&location=Berlin%2C%20Berlin%2C%20Germany&geoId=106967730&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0'

driver.implicitly_wait(10)
driver.get(url1)

n = driver.find_elements(By.CLASS_NAME, "results-context-header__job-count")[0].text
# y=pd.to_numeric(n)
# y=pd.to_numeric(n)
```

```python
#Job scrolling code:
# i=1
# while i<int((n+200)/25):
#     driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
#     i=i+1

wait = WebDriverWait(driver, 15)
job_list = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "jobs-search__results-list")))
jobs = job_list.find_elements(By.TAG_NAME, "li")

for job in jobs:
    try:
        # Job title
        title = job.find_element(By.CLASS_NAME, "base-search-card__title").text.strip()

        # Company name
        company = job.find_element(By.CLASS_NAME, "base-search-card__subtitle").text.strip()

        # Apply link (job card usually has an <a>)
        link = job.find_element(By.TAG_NAME, "a").get_attribute("href")

        # Company logo (if available)
        try:
            logo = job.find_element(By.CLASS_NAME, "artdeco-entity-image").get_attribute("src")
        except: 
            logo = "No logo found"

        print(f"Title: {title}")
        print(f"Company: {company}")
        print(f"Logo: {logo}")
        print(f"Apply link: {link}")
        print("-" * 50)

    except Exception as e:
        print("Error parsing job:", e)

driver.quit()
```
